```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,20,100)
y = np.sin(x)
z = np.sin(x+(np.pi)/2)
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
x = np.linspace(0,20,100)
y = np.cos(x)
plt.plot(x,y,'red')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Waveform")
plt.show()
```

```python
x = np.linspace(0,20,100)
y = np.cos(x)
plt.xlim(-5,25)
plt.ylim(-2,2)
plt.plot(x,y,'red')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Waveform")
plt.show()
```

```python
x = np.linspace(0,20,100)
y = np.sin(x)
z = np.sin(x+(np.pi)/2)
plt.plot(x,y,"red")
plt.plot(x,z,"blue")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Waveform")
plt.legend(["sine","cosine"])
plt.show()
```

```python
x = np.linspace(0,20,50)
y = np.sin(x)
z = np.sin(x+(np.pi)/2)
plt.stem(x,y,"red")
plt.stem(x,z,"blue")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Waveform")
plt.legend(["sine","cosine"])
plt.show()
```

```python
data = pd.read_csv("C:\\Users\\fy.SIESGST\\Downloads\\countries.csv")
us = data[data.country=='United States']
china = data[data.country=='China']
plt.plot(us.year,us.population/10**9,'green',label="United States")
plt.plot(china.year,china.population/10**9,'blue',label="China")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
us_growth_rate=(us.population-us.population.min())/us.population.min()
china_growth_rate=(china.population-china.population.min())/china.population.min()
plt.plot(us.year,us_growth_rate,'o')
plt.plot(china.year,china_growth_rate,'*')
plt.show()
```

```python
growth_rate=(us.population-us.population.iloc[0])/us.population.iloc[0]
plt.plot(us.year,growth_rate)
plt.show()
```
